# Лабораторная работа 1

№1

y'=kx
k=0.42

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dxdt(x,t):
    k=0.42
    return k * x

t = np.linspace(0, 10, 100)
```

```python
solve = odeint(dxdt, y0=1, t=t)
solve = solve.flatten()
```

```python
plt.plot(t, solve, label = r"y'=kx")
plt.grid(True)
plt.legend()
plt.show()
```

№2

y'=k(x)*x
k(x)=ax-b
a=1,b=1

```python
def k(x):
    return 1*x-1
def dxdt(x, t):
    return k(x) * x


t = np.linspace(0, 10, 100) 
```

```python
solve = odeint(dxdt, y0=1, t=t)
solve = solve.flatten()
```

```python

plt.plot(t, solve, label = r"y'=k(x)*x")
plt.grid(True)
plt.legend()
plt.show()
```

№3

y'=k(x)*x-c(x) k(x)=ax-b a=1,b=2 c(x)=f*x f=2

```python
def k(x):
    return 1*x-2
def c(x):
    return 2*x
def dxdt(x, t):
    return k(x) * x - c(x)

t = np.linspace(0, 10, 100) 
```

```python
solve = odeint(dxdt, y0=1, t=t)
solve = solve.flatten()
```

```python
plt.plot(t, solve, label = r"y'=k(x)*x-c(x)")
plt.grid(True)
plt.legend()
plt.show()
```
